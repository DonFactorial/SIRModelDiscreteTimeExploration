```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rhs(y, beta, gamma):
    return np.array([y[0] - beta*y[0]*y[1] + gamma*y[1],
                     y[1] + beta*y[0]*y[1] - gamma*y[1]])
    
beta = 0.1
gamma1 = 3
gamma2 = 6
init_cond = np.array([49, 1])
t_max = 100
t_vals = np.arange(0,t_max+1)
SI_vals1 = np.zeros((2, t_vals.size))
SI_vals1[:,0] = init_cond
SI_vals2 = np.zeros_like(SI_vals1)
SI_vals2[:,0] = init_cond
for i in range(1, t_vals.size):
    SI_vals1[:,i] = rhs(SI_vals1[:,i-1], beta, gamma1)
    SI_vals2[:,i] = rhs(SI_vals2[:,i-1], beta, gamma2)
    
fig, axs = plt.subplots(1, 2, figsize=(10,4))
fig.suptitle(f"SI Model with $N={np.sum(init_cond)}$")
axs[0].plot(t_vals, SI_vals1[0,:], label="Susceptible")
axs[0].plot(t_vals, SI_vals1[1,:], label="Infected")
axs[0].set_xlabel("Time")
axs[0].set_ylabel("Population")
axs[0].set_title(f"$b={beta}$, $g={gamma1}$")

axs[1].plot(t_vals, SI_vals2[0,:], label="Susceptible")
axs[1].plot(t_vals, SI_vals2[1,:], label="Infected")
axs[1].set_xlabel("Time")
axs[1].set_ylabel("Population")
axs[1].set_title(f"$b={beta}$, $g={gamma2}$")
axs[1].legend()
plt.show()
```

```python
def rhs2(y, beta, gamma):
    return np.array([y[0] - beta*y[0]*y[1],
                     y[1] + beta*y[0]*y[1] - gamma*y[1],
                     y[2] + gamma*y[1]])
    
beta = 0.1
gamma1 = 0.5
gamma2 = 3
init_cond = np.array([17, 3, 0])
t_max = 20
t_vals = np.arange(0,t_max+1)
SIR_vals1 = np.zeros((3, t_vals.size))
SIR_vals1[:,0] = init_cond
SIR_vals2 = np.zeros_like(SIR_vals1)
SIR_vals2[:,0] = init_cond
for i in range(1, t_vals.size):
    SIR_vals1[:,i] = rhs2(SIR_vals1[:,i-1], beta, gamma1)
    SIR_vals2[:,i] = rhs2(SIR_vals2[:,i-1], beta, gamma2)
    
fig, axs = plt.subplots(1, 2, figsize=(10,4))
fig.suptitle(f"SIR Model with $N={np.sum(init_cond)}$")
axs[0].plot(t_vals, SIR_vals1[0,:], label="Susceptible")
axs[0].plot(t_vals, SIR_vals1[1,:], label="Infected")
axs[0].plot(t_vals, SIR_vals1[2,:], label="Recovered")
axs[0].set_xlabel("Time")
axs[0].set_ylabel("Population")
axs[0].set_title(f"$b={beta}$, $g={gamma1}$")

axs[1].plot(t_vals, SIR_vals2[0,:], label="Susceptible")
axs[1].plot(t_vals, SIR_vals2[1,:], label="Infected")
axs[1].plot(t_vals, SIR_vals2[2,:], label="Recovered")
axs[1].set_xlabel("Time")
axs[1].set_ylabel("Population")
axs[1].set_title(f"$b={beta}$, $g={gamma2}$")
axs[1].legend()
plt.show()
```
